```python
# Load dataset

df = pd.read_csv('Oil and Gas 1932-2014.csv')
df.head()
```

```python
df[['oil_prod32_14', 'oil_price_2000', 'oil_value_nom', 'population']].isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Analyse global production over time
df_grouped = df.groupby('year')['oil_prod32_14'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_grouped['year'], df_grouped['oil_prod32_14'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Oil Production')
plt.title('Oil Production Over Time')
plt.grid(True)
plt.show()
```

```python
#Gas Production Over Time

df_grouped_gas = df.groupby('year')['gas_prod55_14'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_grouped_gas['year'], df_grouped_gas['gas_prod55_14'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Gas Production')
plt.title('Gas Production Over Time')
plt.grid(True)
plt.show()
```

```python
#Oil Exprot Over Time
df_oil_exports = df.groupby('year')['oil_exports'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_oil_exports['year'], df_oil_exports['oil_exports'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Oil Exports')
plt.title('Oil Exports Over Time')
plt.grid(True)
plt.show()
```

```python
# Gas Exports Over Time

df_gas_exports = df.groupby('year')['gas_exports'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(df_gas_exports['year'], df_gas_exports['gas_exports'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Gas Exports')
plt.title('Gas Exports Over Time')
plt.grid(True)
plt.show()
```

```python
df_gas_exports.head()
```

```python
# Top Oil Producers In 2010

df_2010 = df[df['year'] == 2010]
top_producers = df_2010[["cty_name", "oil_prod32_14"]].dropna().sort_values(by="oil_prod32_14", ascending=False).head(10)

top_producers.plot(x="cty_name", y="oil_prod32_14", kind="bar", figsize=(10, 6))
plt.xlabel("Country")
plt.ylabel("Oil Production")
plt.title("Top Producers In 2010")
plt.show()
```

```python
# Relationship between Oil Price and Oil Production

plt.figure(figsize=(8,5))
plt.scatter(df['oil_price_2000'], df['oil_prod32_14'], alpha=0.3)
plt.xlabel('Oil Price (USD)')
plt.ylabel('Oil Production')
plt.title('Oil Price vs Production')
plt.grid(True)
plt.show()
```

```python
# Relationship between Gas Prices and Gas Production

plt.figure(figsize=(8,5))
plt.scatter(df['gas_price_2000'], df['gas_prod55_14'], alpha=0.3)
plt.xlabel('Gas Price (USD)')
plt.ylabel('Gas Production')
plt.title('Gas Price vs Production')
plt.grid(True)
plt.show()
```

```python
# Top Gas Producers in 2010

df_gas_2010 = df[df['year'] == 2010]
top_gas_prod = df_gas_2010[["cty_name", "gas_prod55_14"]].dropna().sort_values(by="gas_prod55_14", ascending=False).head(10)

top_gas_prod
```

```python
top_gas_prod.plot(x="cty_name", y="gas_prod55_14", kind="bar", figsize=(10, 6))
plt.xlabel("Country")
plt.ylabel("Gas Production")
plt.title("Top Gas Producers In 2010")
plt.show()
```

```python
# Lowest Oil producers in 2010

df_lowest_oil = df_2010[["cty_name", "oil_prod32_14"]].dropna().sort_values(by="oil_prod32_14", ascending=True).head(10)

df_lowest_oil
```

```python
# Relationship between Prodution, Price, Export, and Popuation

corr = df[['oil_prod32_14', 'oil_price_2000', 'oil_exports', 'population']].corr()
corr
```

```python
# Plot of Correltion

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Key Variables')
plt.show()
```

```python
# Countries with the Highest growth in Oil Production 1980 - 2010

prod_1980 = df[df['year'] == 1980][['cty_name', 'oil_prod32_14']].set_index('cty_name')
prod_2010 = df[df['year'] == 2010][['cty_name', 'oil_prod32_14']].set_index('cty_name')

growth = prod_2010 - prod_1980
growth = growth.sort_values(by='oil_prod32_14', ascending=False).head(10)
growth
```

```python
# Plot
growth.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Growth in Oil Production')
plt.title('Countries with the Highest Growth in Oil Production')
plt.show()
```

```python
# Countries with Highest Growth in Gas production 1980 - 2010

gas_prod_1980 = df[df['year'] == 1980][['cty_name', 'gas_prod55_14']].set_index('cty_name')
gas_prod_2010 = df[df['year'] == 2010][['cty_name', 'gas_prod55_14']].set_index('cty_name')

gas_growth = gas_prod_2010 - gas_prod_1980
gas_growth = gas_growth.sort_values(by='gas_prod55_14', ascending=False).head(10)
gas_growth
```

```python
# Plot
gas_growth.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Growth in Gas Production')
plt.title('Countries with the Highest Growth in Gas Production')
plt.show()
```

```python
# Top Oil Exporting Countires

df_top_oil_exports = df.groupby('cty_name')['oil_exports'].sum().reset_index()
df_top_oil_exports = df_top_oil_exports.sort_values(by='oil_exports', ascending=False).head(10)
df_top_oil_exports
```

```python
# Plot

df_top_oil_exports.plot(x='cty_name', y='oil_exports', kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Oil Exports')
plt.title('Top Oil Exporting Countries')
plt.show()
```

```python
# Top Gas Exporting Countries

df_top_gas_exports = df.groupby('cty_name')['gas_exports'].sum().reset_index()
df_top_gas_exports = df_top_gas_exports.sort_values(by='gas_exports', ascending=False).head(10)
df_top_gas_exports
```

```python
# Plot

df_top_gas_exports.plot(x='cty_name', y='gas_exports', kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Gas Exports')
plt.title('Top Gas Exporting Countries')
plt.show()
```

```python
# Countries with High Population

df_high_pop = df.groupby('cty_name')['population'].sum().reset_index()
df_high_pop = df_high_pop.sort_values(by='population', ascending=False).head(10)
df_high_pop
```

```python
# Plot

df_high_pop.plot(x='cty_name', y='population', kind='bar', figsize=(10, 6))
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Countries with High Population')
plt.show()
```
